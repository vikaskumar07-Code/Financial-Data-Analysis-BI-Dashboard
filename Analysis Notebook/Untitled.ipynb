## Missing Value Classification
Low Missing (< 10%) :- Usually manageable.
action :- Investigate, Recalculate if possible, Otherwise preserve

Medium Missing (10%–30%) :- Need business justification.
action :- Column-by-column decision

High Missing (> 30%) :- Very careful.
action :- Usually preserve as NaN, Don't force imputation,Evaluate business importance

```python
missing_reports = {}

for name, df in datasets.items():

    report = pd.DataFrame({
        'missing_count': df.isnull().sum(),
        'missing_percentage': round((df.isnull().sum()/len(df))*100, 2)
    })

    report = (
        report
        .query("missing_count > 0")
        .sort_values("missing_percentage", ascending=False)
    )

    missing_reports[name] = report
```

During KPI development, certain ratio calculations produced infinite values because a small number of companies had zero sales, zero assets, or zero equity. Instead of removing those companies or imputing artificial values, the infinite results were converted to missing values (NaN), preserving the integrity of the original financial data while ensuring accurate KPI analysis.

Not removing the outliers the objective of the project was corporate benchmarking across the entire listed universe. Extreme values represented genuine business conditions such as distressed companies, high-growth firms, and highly leveraged organizations. Removing them would have distorted the real financial landscape. Therefore, outliers were retained while using median-based interpretation where appropriate.
